```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
flight_file = "gov2016_flight_delay.csv"
flight_df = pd.read_csv(flight_file)
flight_df.head()
```

```python
#look for missing values 
flight_df.count()
```

```python
#remove the rows with missing data
clean_flight_df = flight_df.dropna(how="any")
clean_flight_df.count()
```

```python
#filter the data so that only those sighting in the US are in a data frame
AA_flight_df = clean_flight_df.loc[clean_flight_df["carrier"] =="AA", :]
AA_flight_df
```

```python
AA_counts = AA_flight_df ["weather_delay"].value_counts()
AA_counts
```

```python
#convert the count to series into dataframe 
AA_flight_counts_df = pd.DataFrame(AA_counts)
AA_flight_counts_df.head()

```
